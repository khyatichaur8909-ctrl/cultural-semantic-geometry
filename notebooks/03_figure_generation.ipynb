```python
import pandas as pd
import matplotlib.pyplot as plt

# ==========================
# LOAD DATA
# ==========================

tiny = pd.read_csv("tinyllama_control_summary.csv")
qwen = pd.read_csv("qwen_control_summary.csv")

# ==========================
# PUBLICATION SETTINGS
# ==========================

plt.rcParams.update({
    "font.size": 11,
    "axes.labelsize": 11,
    "axes.titlesize": 12,
    "legend.fontsize": 9
})

# ==========================
# CREATE FIGURE
# ==========================

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# --------------------------
# TinyLlama
# --------------------------

axes[0].plot(
    tiny["layer"],
    tiny["dist_asym"],
    linewidth=2,
    label="J_asym"
)

axes[0].plot(
    tiny["layer"],
    tiny["dist_lit"],
    linewidth=2,
    label="E_lit"
)

axes[0].plot(
    tiny["layer"],
    tiny["dist_sym"],
    linewidth=2,
    label="S_sym"
)

axes[0].plot(
    tiny["layer"],
    tiny["dist_scram"],
    linewidth=2,
    label="J_scram"
)

axes[0].set_title("TinyLlama-1.1B")
axes[0].set_xlabel("Transformer Layer")
axes[0].set_ylabel("Cosine Distance")
axes[0].grid(alpha=0.3)

# --------------------------
# Qwen
# --------------------------

axes[1].plot(
    qwen["layer"],
    qwen["dist_asym"],
    linewidth=2,
    label="J_asym"
)

axes[1].plot(
    qwen["layer"],
    qwen["dist_lit"],
    linewidth=2,
    label="E_lit"
)

axes[1].plot(
    qwen["layer"],
    qwen["dist_sym"],
    linewidth=2,
    label="S_sym"
)

axes[1].plot(
    qwen["layer"],
    qwen["dist_scram"],
    linewidth=2,
    label="J_scram"
)

axes[1].set_title("Qwen-1.8B")
axes[1].set_xlabel("Transformer Layer")
axes[1].grid(alpha=0.3)

# --------------------------
# LEGEND
# --------------------------

axes[1].legend(loc="best")

# --------------------------
# FINAL LAYOUT
# --------------------------

plt.tight_layout()

# ==========================
# SAVE FIGURES
# ==========================

plt.savefig(
    "Figure1_Trajectories.png",
    dpi=600,
    bbox_inches="tight"
)

plt.savefig(
    "Figure1_Trajectories.pdf",
    bbox_inches="tight"
)

plt.show()
```
